7. Utilizando avgMins y avgOutCalls:
###### a) Normalizar usando como valor maximo 2 desviaciones estandar de la media en nuevas columnas. Guardar los resultado en nuevas columna, avgMinsNorm y avgOutCallsNorm.

```python
def normalize(minimun, std, value):
    return (value-minimun)/(2*std - minimun)

telephone_company_df['avgOutCallsNorm'] = telephone_company_df.apply (lambda row: normalize(telephone_company_df.avgOutCalls.min(), telephone_company_df.avgOutCalls.std(), row.avgOutCalls),axis=1)
telephone_company_df['avgMinsNorm'] = telephone_company_df.apply (lambda row: normalize(telephone_company_df.avgMins.min(), telephone_company_df.avgMins.std(), row.avgMins),axis=1)

telephone_company_df[['avgMinsNorm', 'avgOutCallsNorm']]
```

###### d) Usando la función apply() de pandas, discretizar la columna avgMins en 5 bins distintos usando:
1) Igual Frecuencia

```python
number_of_bins = 5
```

```python
def equal_frecuency(feature_column, bins_nr):
    bin_size = feature_column.count()/bins_nr
    sorted_column = feature_column.sort_values().to_frame()
    sorted_column['bin'] = np.concatenate([index*np.ones(shape=bin_size, dtype=np.int) for index in np.arange(bins_nr)])
    return sorted_column
    
equal_frecuency(telephone_company_df['avgMins'], number_of_bins)
```

2) igual ancho

```python
number_of_bins = 5
```

```python
def equal_width(feature_column, bins_nr):
    feature_range = feature_column.max() - feature_column.min()
    bin_width = feature_range/bins_nr
    intervals = np.arange(start=0, stop=feature_range+1, step=bin_width)
    equal_width_df = feature_column.to_frame()
    equal_width_df['bin'] = pd.cut(feature_column.to_frame()['avgMins'], bins=intervals, include_lowest=True)
    return equal_width_df

equal_width(telephone_company_df['avgMins'],number_of_bins)
```

### Problema 3
Cargar el dataset (ACMETelephoneABT.csv) en un Dataframe de Spark. Usando Spark Pipelines, limpiar los atributos: avgDroppedCalls, creditCard, handsetAge, marriageStatus, numHandsets, regionType forma que todos ellos sean numericos y normalizados. Utilizar one hot encoding para transformar atributos categoricos en numericos.

```python
import pyspark
sc = pyspark.SparkContext(master='local[*]')
```

```python
telephone_company_sdf = sc.textFile(path).map(lambda x : x.split(","))
header = telephone_company_sdf.first()
telephone_company_sdf = telephone_company_sdf.filter(lambda x : x!=header).map(lambda x : [float(x[26]), str(x[13]), float(x[8]), str(x[4]), float(x[7]), str(x[3]), str(x[-1])])
```

```python
# Haciendo unique en regionType se obtienen las siguientes categorias categorias:
#' town', ' ', ' suburban', ' rural', ' s', ' unknown', ' t', ' r'

# Tengo que hacer el cleaning de tal forma que town == t, suburban == s, rural == r, unknown == ' '

def regionType_cleaner(x):
    regionType = x[5]
    if regionType ==' town' or regionType ==' t':
        x[5] = 'town'
    elif regionType ==' suburban' or regionType ==' s':
        x[5] = 'suburban'
    elif regionType ==' rural' or regionType ==' r':
        x[5] = 'rural'
    else:
        x[5] = 'unknown'
    return x

telephone_company_sdf = telephone_company_sdf.map(regionType_cleaner)
```

Es necesario entonces tratar con los datos unknowns. debido a que corresponden a ser missing values

```python
missingValues = telephone_company_sdf.filter(lambda x : x[5]=='unknown').count()
towns = telephone_company_sdf.filter(lambda x : x[5]=='town').count()
suburbans = telephone_company_sdf.filter(lambda x : x[5]=='suburban').count()
rurals = telephone_company_sdf.filter(lambda x : x[5]=='rural').count()
(missingValues, towns, suburbans, rurals)
```

```python
total_regionTypes = telephone_company_sdf.count()
missing_values_per = (missingValues*100)/total_regionTypes
print 'Missing values '+str(missing_values_per)+'%'
towns_per = (towns*100)/total_regionTypes
print 'Towns '+str(towns_per)+'%'
suburbans_per = (suburbans*100)/total_regionTypes
print 'Suburbans '+str(suburbans_per)+'%'
rurals_per = (rurals*100)/total_regionTypes
print 'Missing values '+str(rurals_per)+'%'
```

Como se puede ver los missing values representan el 48% de los datos. Esto es mucho mayor que los de las categorias restantes. El libro sugiere que para un alto porcentaje de missing values se utilize el 'rule of thumb' la cual sugiere que si un feature tiene un 60% de missing values directamente no se considere, y que a partir del 30% de missing values realizar un promedio entre las categorias restantes no llevaria a resultados efectivos. La unica solucion seria realizar un modelo el cual prediga cual seria este valor. Pero dado a que esto puede ser considerado como un overkill para el obligatorio simplemente ignoraremos este feature.

##### Cleaning del Feature creditCard

```python
# Haciendo unique en creditCard se obtienen las siguientes categorias categorias:
#' true', ' false', ' no', ' t', ' f', ' yes'

# Tengo que limpiarlo de tal forma que true == t == yes, false == f === n

def creditCard_cleaner(x):
    creditCard = x[1]
    if creditCard ==' true' or creditCard ==' t' or creditCard == ' yes':
        x[1] = 'True'
    else:
        x[1] = 'False'
    return x

telephone_company_sdf = telephone_company_sdf.map(creditCard_cleaner)
```

Por lo tanto tampoco considero utilizar esta feature.

```python
from pyspark.sql import SQLContext, Row
sqlContext = SQLContext(sc) # sino no tengo toDF();

telephone_df = telephone_company_sdf.map(lambda x : Row(avgDroppedCalls = x[0],
                                 creditCard = x[1],
                                 handsetAge = x[2],
                                 numHandsets = x[4],
                                 churn = x[6])).toDF()
```

```python
telephone_df.show()
```

##### Transformar las features categoricas y luego las normalizo

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler, StandardScaler
from pyspark.ml import Pipeline
from pyspark.mllib.linalg import Vectors
from pyspark.mllib.regression import LabeledPoint

# creditCard encoding
creditCardIndexer = StringIndexer(inputCol='creditCard', outputCol='creditCard_idx')
creditCardEncoder = OneHotEncoder(dropLast=False, inputCol="creditCard_idx", outputCol="creditCardVec")

# chrun indexing
churnIndexer = StringIndexer(inputCol='churn', outputCol='label')

# Vector assembling
vectorAssembling = VectorAssembler(inputCols=['avgDroppedCalls', 'handsetAge', 'numHandsets', 'creditCardVec'],
                                   outputCol="all_features")

# Feature scaling
featureScaling = StandardScaler(inputCol='all_features', outputCol='features')

stages = [creditCardIndexer, creditCardEncoder, churnIndexer, vectorAssembling, featureScaling]

model = Pipeline(stages=stages).fit(telephone_df)
scaledData = model.transform(telephone_df)

scaledData.show()
labeled = scaledData.select('features','label')
labeled.show()
 
```

##### 2. Entrene arboles de decision modificando en cada uno de ellos los par ́ametros: max-Depth=[1, 5, 10], maxBins=[16, 32, 64] y impurity=[’gini’, ’entropy’]. 

Determine cual es el modelo que tiene mejor precisi ́on en detectar churn (porcentaje de veces en que los clientes con churn son identificados como tales en el test dataset). Utilice un split de 70 % para entrenar los  ́arboles y el resto para evaluarlos.

```python
label_indexer = StringIndexer(inputCol="label", outputCol="label_idx").fit(labeled)
```

```python
from pyspark.ml.classification import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(labelCol="label_idx", featuresCol="features")
```

```python
stages = [label_indexer, decision_tree]
pipeline = Pipeline(stages=[label_indexer, decision_tree])

(training_data, test_data) = labeled.randomSplit([0.7, 0.3])

model = pipeline.fit(training_data)
```

```python
model.transform(test_data).select('label', 'prediction').show()
```
